```python
upper_limit = df4['new_cases'].mean() + 3*df4['new_cases'].std()
lower_limit = df4['new_cases'].mean() - 3*df4['new_cases'].std()
print('upper limit is ', upper_limit)
print('lower limit is ', lower_limit)
```

```python
up_limit = df4['new_deaths'].mean() + 3*df4['new_deaths'].std()
lw_limit = df4['new_deaths'].mean() - 3*df4['new_deaths'].std()
print('upper limit of death case is ', up_limit)
print('lower limit of death is ', lw_limit)
```

```python
df4.loc[(df4['new_cases'] > upper_limit) | (df4['new_cases'] < lower_limit)]
```

```python
df4.loc[(df4['new_deaths'] > up_limit) | (df4['new_deaths'] < lw_limit)]
```

```python
#outlier removal of new cases z score method
new_df = df4.loc[(df4['new_cases']< upper_limit) & (df4['new_cases'] > lower_limit)]
print('Before outliers removal ',len(df4))
print('After outliers removal ',len(new_df))
print('outliers ', len(df4)-len(new_df))
```

```python
#outlier removal of death cases z score method
new_df = df4.loc[(df4['new_deaths']< up_limit) & (df4['new_deaths'] > lw_limit)]
print('Before outliers removal ',len(df4))
print('After outliers removal ',len(new_df))
print('outliers ', len(df4)-len(new_df))
```

```python
arimacases = new_df.drop(["population_density","extreme_poverty","human_development_index","temp",	"dew",	"humidity",	"windspeed"], axis = 1)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def ad_test(dataset):
    arimacasestest = adfuller(dataset, autolag = 'AIC')
    print("1. ADF : ",arimacasestest[0])
    print("2. p-value : ",arimacasestest[1])
    print("3. no.lags : ",arimacasestest[2])
    print("4. no. observation for adf regression and critical value : ",arimacasestest[4])
    print("5. critical values : ")
    for key, val in arimacasestest[4].items():
        print("\t",key, ": ",val)
```

```python
ad_test(arimacases['new_cases'])
```
